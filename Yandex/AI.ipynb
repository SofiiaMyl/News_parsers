1. Рабочий web_search

```python
import openai
import json 

region="213" # Москва
# region="10176" # Ненецкий АО

client = openai.OpenAI(
    api_key=YANDEX_CLOUD_API_KEY,
    base_url="https://ai.api.cloud.yandex.net/v1",
    project=YANDEX_CLOUD_FOLDER
)

response = client.responses.create(
    model=f"gpt://{YANDEX_CLOUD_FOLDER}/{YANDEX_CLOUD_MODEL}",
    input="""Найди ВСЕ новости, касающиеся компании Севержилкомсервис МП ЗР, СЖКС за 2025 год.

    В ответе выведи ТОЛЬКО json-файл с ТОЧНОЙ структурой:

    {
        'new':{
            'CompanyName': {company name},
            'NewsTitle': {title},
            'NewsDate': {date},
            'NewsSource': {domain where was found the article},
            'NewsURL': {full url of the article},
            'NewsText': {full text of article},
        }
    }
    """,
    tools=[
        {
           "type": "web_search",
           "filters": {
               "allowed_domains": [
                   "adm-nao.ru", "nao24.ru", "nvinder.ru", "www.info83.ru"
               ]
           },
           "user_location": {
                   "region": region,
               },
        "search_context_size": "high", # варианты: low | medium | high
        }
    ],
    temperature=0.1,
    max_output_tokens=100000000
)

# Текст ответа
print("Текст ответа:")
print(response.output_text)
# print("\n" + "=" * 50 + "\n")

# Полный ответ
# print("Полный ответ (JSON):")
# print(json.dumps(response.model_dump(), indent=2, ensure_ascii=False))
```
